### NULL값을 삭제할지 대체할지 확인하기 위해 시각화

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

data = pd.read_csv('C:/Users/박휘석/Desktop/고급비즈니스애널리틱스/PurchasesFINAL12312016-1.csv')


# 'Size' 열의 단위를 ml로 통일하는 함수
def convert_to_ml(size_str):
    size_str = str(size_str).lower().replace(' ', '')
    
    if 'ml' in size_str:
        value = re.findall(r'(\d+\.?\d*)', size_str)
        if value:
            return float(value[0])
    elif 'l' in size_str:
        value = re.findall(r'(\d+\.?\d*)', size_str)
        if value:
            return float(value[0]) * 1000
    elif 'oz' in size_str:
        value = re.findall(r'(\d+\.?\d*)', size_str)
        if value:
            return float(value[0]) * 29.5735

    return None

# 'data' 데이터프레임의 'Size' 열에 함수 적용
data['Size_ml'] = data['Size'].apply(convert_to_ml)

# 히스토그램 다시 그리기
plt.figure(figsize=(10, 6))
plt.hist(data['Size_ml'].dropna(), bins=1000, edgecolor='black', color='skyblue')
plt.title('Histogram of Size (ml)', fontsize=15)
plt.xlabel('Size (ml)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.75)
plt.xlim(0, 6000)  # x축 범위를 0에서 20000ml로 제한
plt.show()
```

### NULL값을 중앙값으로 대체

```python
# 'Size_ml' 열의 중앙값 계산
size_median = data['Size_ml'].median()

# 'Size_ml' 열의 널값을 중앙값으로 대체
data['Size_ml'].fillna(size_median, inplace=True)

# 널값 대체 후 'Size_ml' 열의 널값 개수를 확인하여 제대로 처리되었는지 검증
null_count_after = data['Size_ml'].isnull().sum()

print(f"Size_ml 열의 중앙값: {size_median}")
print(f"널값 대체 후 Size_ml 열의 널값 개수: {null_count_after}")
```

```python
import pandas as pd

# 'PODate'를 datetime 형식으로 변환
data['PODate'] = pd.to_datetime(data['PODate'])

# 날짜와 상점별로 그룹화하여 일별 총 매출 계산
daily_store_sales = data.groupby([data['PODate'].dt.date, 'Store'])['Dollars'].sum().reset_index()

# 컬럼명 정리 및 데이터 타입 변환
daily_store_sales.rename(columns={'PODate': 'PODate_daily'}, inplace=True)
daily_store_sales['PODate_daily'] = pd.to_datetime(daily_store_sales['PODate_daily'])

# 날짜 순으로 정렬
daily_store_sales.sort_values(by='PODate_daily', inplace=True)

# 결과 확인 (선택 사항)
print(daily_store_sales.head())
```

```python
import matplotlib.pyplot as plt

# 판매량이 많은 상위 5개 상점만 분석
top5_stores = daily_store_sales['Store'].value_counts().nlargest(5).index

# 롤링 평균 계산 및 시각화
plt.figure(figsize=(15, 8))

# 각 상점별 롤링 평균 계산 및 플로팅
for store_id in top5_stores:
    store_data = daily_store_sales[daily_store_sales['Store'] == store_id].copy()
    
    # 7일 롤링 평균 계산
    store_data['RollingMeanDollars'] = store_data['Dollars'].rolling(window=7, min_periods=1).mean()
    
    # 선 그래프 그리기
    plt.plot(store_data['PODate_daily'], store_data['RollingMeanDollars'], label=f'Store: {store_id}')

# 제목과 라벨 추가
plt.title('7-Day Rolling Average of Total Daily Dollars by Store')
plt.xlabel('Date')
plt.ylabel('Rolling Mean Dollars')
plt.legend(title='Store')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# 예시 데이터프레임 (df)가 이미 로드되어 있다고 가정

# 1. 전체 기간 동안의 총 매출을 기준으로 가장 높은/낮은 스토어 찾기
total_sales_by_store = df.groupby('Store')['Dollars'].sum()
top_store_id = total_sales_by_store.idxmax()
lowest_store_id = total_sales_by_store.idxmin()

# 2. 가장 매출이 높은 스토어의 모든 거래 내역 필터링 및 상품 요약
top_store_data = df[df['Store'] == top_store_id].copy()
top_product_summary = top_store_data.groupby('Description').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalDollars=('Dollars', 'sum')
).reset_index()
top_product_summary['AverageUnitPrice'] = top_product_summary.apply(
    lambda row: row['TotalDollars'] / row['TotalQuantity'] if row['TotalQuantity'] > 0 else 0, axis=1
)
top_10_products = top_product_summary.sort_values(by='TotalQuantity', ascending=False).head(10)
sorted_top_10 = top_10_products.sort_values(by='TotalQuantity', ascending=True)

# 3. 가장 매출이 낮은 스토어의 모든 거래 내역 필터링 및 상품 요약
lowest_store_data = df[df['Store'] == lowest_store_id].copy()
lowest_product_summary = lowest_store_data.groupby('Description').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalDollars=('Dollars', 'sum')
).reset_index()
lowest_product_summary['AverageUnitPrice'] = lowest_product_summary.apply(
    lambda row: row['TotalDollars'] / row['TotalQuantity'] if row['TotalQuantity'] > 0 else 0, axis=1
)
lowest_10_products = lowest_product_summary.sort_values(by='TotalQuantity', ascending=False).head(10)
sorted_lowest_10 = lowest_10_products.sort_values(by='TotalQuantity', ascending=True)

# 4. 두 그래프를 위한 서브플롯 생성
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# 5. 첫 번째 그래프 (매출 1위 스토어) 그리기
norm_top = plt.Normalize(sorted_top_10['AverageUnitPrice'].min(), sorted_top_10['AverageUnitPrice'].max())
colors_top = cm.Greens(norm_top(sorted_top_10['AverageUnitPrice']))
ax1.barh(sorted_top_10['Description'], sorted_top_10['TotalQuantity'], color=colors_top, edgecolor='black')
ax1.set_title(f'Top 10 Products by Quantity for Store of Highest income\n(Color Intensity by Average Unit Price)', fontsize=16)
ax1.set_xlabel('Total Quantity', fontsize=12)
ax1.set_ylabel('Product Description', fontsize=12)
ax1.grid(axis='x', linestyle='--', alpha=0.6)

# 6. 두 번째 그래프 (매출 꼴찌 스토어) 그리기
norm_lowest = plt.Normalize(sorted_lowest_10['AverageUnitPrice'].min(), sorted_lowest_10['AverageUnitPrice'].max())
colors_lowest = cm.Oranges(norm_lowest(sorted_lowest_10['AverageUnitPrice']))
ax2.barh(sorted_lowest_10['Description'], sorted_lowest_10['TotalQuantity'], color=colors_lowest, edgecolor='black')
ax2.set_title(f'Top 10 Products by Quantity for Store lowest income\n(Color Intensity by Average Unit Price)', fontsize=16)
ax2.set_xlabel('Total Quantity', fontsize=12)
ax2.set_ylabel('Product Description', fontsize=12)
ax2.grid(axis='x', linestyle='--', alpha=0.6)

# 7. 전체 레이아웃 조정 및 시각화
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 상관관계 행렬 확인
# 세 변수만 선택해서 상관관계 행렬을 계산해
correlation_matrix = df[['PurchasePrice', 'Quantity', 'Dollars']].corr()
print("## 1. PurchasePrice, Quantity, Dollars 간의 상관관계 행렬")
print(correlation_matrix)
print("-" * 50)


# 2. Dollars가 PurchasePrice와 Quantity의 곱과 일치하는지 확인
# 'Dollars_calculated'라는 새로운 열을 만들어 곱셈 결과를 저장해
df['Dollars_calculated'] = df['PurchasePrice'] * df['Quantity']

# Dollars와 Dollars_calculated가 정확히 일치하는 행의 개수를 세
exact_match_count = (df['Dollars'] == df['Dollars_calculated']).sum()

# 소수점 차이 때문에 정확히 일치하지 않는 경우도 있으므로
# 차이가 0.01 미만인 경우도 확인해볼 수 있어.
# (abs(df['Dollars'] - df['Dollars_calculated']) < 0.01).sum()

total_rows = len(df)
print("## 2. Dollars와 (PurchasePrice * Quantity) 일치 여부")
print(f"전체 거래 수: {total_rows:,}개")
print(f"Dollars가 계산된 값과 정확히 일치하는 거래 수: {exact_match_count:,}개")
print(f"일치율: {(exact_match_count / total_rows) * 100:.2f}%")
print("-" * 50)


# 3. 산점도(Scatter Plot) 시각화
plt.style.use('seaborn-v0_8-whitegrid') # 그래프 스타일 설정
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 첫 번째 산점도: PurchasePrice vs. Dollars
axes[0].scatter(df['PurchasePrice'], df['Dollars'], alpha=0.5, color='royalblue', s=10)
axes[0].set_title('PurchasePrice vs. Dollars', fontsize=15)
axes[0].set_xlabel('PurchasePrice', fontsize=12)
axes[0].set_ylabel('Dollars', fontsize=12)

# 두 번째 산점도: Quantity vs. Dollars
axes[1].scatter(df['Quantity'], df['Dollars'], alpha=0.5, color='forestgreen', s=10)
axes[1].set_title('Quantity vs. Dollars', fontsize=15)
axes[1].set_xlabel('Quantity', fontsize=12)
axes[1].set_ylabel('Dollars', fontsize=12)

# 그래프 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 1. 'ReceivingDate'와 'PODate' 열을 datetime 형식으로 변환
# 데이터가 로드된 DataFrame의 변수명은 'df'라고 가정합니다.
df['ReceivingDate'] = pd.to_datetime(df['ReceivingDate'], errors='coerce')
df['PODate'] = pd.to_datetime(df['PODate'], errors='coerce')

# 2. 배송 기간(일)을 계산하는 새로운 컬럼 생성
# ReceivingDate에서 PODate를 빼서 날짜 간 차이를 계산합니다.
df['DeliveryDays'] = (df['ReceivingDate'] - df['PODate']).dt.days

# 3. 배송 기간의 분포를 히스토그램으로 시각화
plt.figure(figsize=(10, 6))
#dropna()를 사용해 결측값을 제외하고, bins를 설정하여 1일 단위로 구간을 나눕니다.
plt.hist(df['DeliveryDays'].dropna(), bins=range(0, 31), color='skyblue', edgecolor='black')
plt.title('Distribution of Delivery Days', fontsize=16)
plt.xlabel('Delivery Days', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(range(0, 31, 2))  # x축 눈금을 2일 단위로 설정하여 가독성을 높입니다.
plt.tight_layout()
plt.xlim(0,16)
plt.show()

# 4. 배송 기간 통계 요약 정보 출력
# describe()를 사용해 평균, 표준편차, 최솟값, 최댓값 등을 요약해서 보여줍니다.
delivery_days_summary = df['DeliveryDays'].describe()
print("## 배송 기간 통계 요약")
print(delivery_days_summary)
```
